```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import colors
import json
```

```python
## filename = 'methanogenesis.json'
with open(filename) as json_file:
    data = json.load(json_file)
tempK, pressb, pLogK, pDelG = data # C, bar
tK = np.array(tempK) + 273.15 # to kelvins
pkb = np.array([i*1e-3 for i in pressb]) # to kilobars
logK = np.array(pLogK)
delG = np.array(pDelG)
#pressB.shape, tempK.shape, logK.shape, delG.shape
#print(presskb.min(), presskb.max())
#print(tempK.min(), tempK.max())
pkb.shape, tK.shape, logK.shape
```

```python
divnorm=colors.TwoSlopeNorm(vmin=np.min(logK), vcenter=0., vmax=np.max(logK))
cp = plt.contourf(tK, pkb, logK, 20, cmap='Spectral', norm=divnorm)
plt.contour(tK, pkb, logK, levels=[0.], linestyles='dashed', colors='black')
plt.colorbar(cp, label='logK')
plt.xlabel('Temperature (K)')
plt.ylabel('Pressure (kB)')
#plt.title('Inorganics - Alanine')
plt.tick_params(which='minor', length=6, width=1.5)
plt.tick_params(length=10, width=1.5)
```
